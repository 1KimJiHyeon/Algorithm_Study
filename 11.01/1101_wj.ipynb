```python
# 최소 최대
n = int(input())
ls = list(map(int, input().split()))
#
# print(min(ls), max(ls))
min_n = 1000001
max_n = -1000001

for i in ls:
    if i < min_n:
        min_n = i
    if i > max_n:
        max_n = i
print(min_n, max_n)
```

```python
# 네 개의 소수
n = int(input())

prime = [None for i in range(n+1)]

for i in range(2, n):
    if prime[i] == None:
        prime[i] = True
        j = i * 2
        while j <= n:
            prime[j] = False
            j += i

prime_nums = [i for i in range(n) if prime[i] ==  True]

def check_four_prime(n):
    total = 0
    for i in prime_nums:
        if i < n:
            for j in prime_nums:
                if i+j < n:
                    for k in prime_nums:
                        if n - i - j - k >= 2:
                            if prime[n - i - j - k] == True:
                                return f'{i} {j} {k} {n - i - j - k}'
    return -1
print(check_four_prime(n))
```

```python
# 미로탐색
n, m = map(int, input().split())
ls = []
for i in range(n):
    ls.append(list(input()))


visited = [[False for _ in range(m)] for __ in range(n)]
cnt_ls = []
check = [[ 1 for _ in range(m)] for __ in range(n)]
dx = (1, 0 ,-1, 0)
dy = (0, 1, 0, -1)

def bfs():
    visited[0][0] = True
    q = [(0,0)]
    while q:
        x, y = q.pop(0)

        for i in range(4):
            next_x = x + dx[i]
            next_y = y + dy[i]

            if next_x >= 0 and next_x < n and next_y >= 0 and next_y < m:
                if ls[next_x][next_y] == '1' and visited[next_x][next_y] == False:
                    check[next_x][next_y] = check[x][y] + 1
                    visited[next_x][next_y] = True
                    q.append((next_x, next_y))

bfs()
print(check[n-1][m-1])
```

```python
# 괄호
n = int(input())
ls = [input() for _ in range(n)]


def check(string):
    for i in range(25):
        string = string.replace('()','')
    if len(string) == 0:
        return 'YES'
    else:
        return 'NO'

# for i in ls:
#     print(check(i))

def check2(string):
    left = 0
    right = 0
    for i in string:
        if i == '(':
            left += 1
        else:
            right += 1
        if right > left:
            return 'NO'
    if left == right:
        return 'YES'
    return 'NO'

for i in ls:
    print(check2(i))
```
