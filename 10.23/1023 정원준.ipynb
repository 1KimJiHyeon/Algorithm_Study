```python
start, end = map(int, input().split())
```

```python
end_rt = int(end**0.5)
ls = [1 for _ in range(end_rt+1)]
ls[0] = 0
ls[1] = 0
for i in range(2, end_rt+1):
    if ls[i] == 1:
        for j in range(i+i, end_rt+1, i):
            ls[j] = 0
```

```python
cnt = 0

for i in range(2, end_rt):
    if ls[i] == 1:
        k = i ** 2
        while k < start:
            k *= i
        while k <= end:
            cnt += 1
#                 print(k, k ** 0.5)
            k *= i

print(cnt)
```

```python
cnt = 0
for i in range(start, end+1):
    for j in range(2, len(ls)):
        if ls[j] == 1:
            k = i
            while k % j == 0:
                print(k,j)
                k /= j
            if k == 1:
                cnt += 1
print(cnt)
```
